```python
class nqueen:
    def __init__(self,n):
        self.n = n
        self.board = [-1]*n
        self.solution = []
        
    def safe(self,row,col):
        for pre_row in range(row):
            pre_col = self.board[pre_row]
            if pre_col == col or abs(pre_col-col)==abs(pre_row-row):
                return False
        return True
    
    def solve_queen(self,row=0):
        if row==self.n:
            self.solution.append(self.board[:])
            return
        for col in range(self.n):
            if self.safe(row,col):
                self.board[row] = col
                self.solve_queen(row+1)
                self.board[row] = -1
                
    def print_sol(self):
        for sol in self.solution:
            for row in range(self.n):
                l = ["." for _ in range(self.n)]
                l[sol[row]] = 'Q'
                print(" ".join(l))
            print('\n'+'-'*(2*self.n-1))
    def solve(self):
        self.solve_queen()
        print(f'total solution for {self.n}-Queen:{len(self.solution)}')
        self.print_sol()
if __name__ =='__main__':
    n = 8
    game = nqueen(n)
    game.solve()
```
