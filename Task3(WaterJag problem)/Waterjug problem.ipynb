```python
class waterjug:
    def __init__(self, jug1, jug2, target):
        self.jug1 = jug1 
        self.jug2 = jug2 
        self.target = target 
        self.visited = set()
    
    def valid_state(self, state):
        return state not in self.visited
    
    def dfs(self, state, path):
        if state in self.visited:
            return False
        
        jug1, jug2 = state
        print(f"Current State: {state}")
        self.visited.add(state)
        path.append(state)
        
        if jug1 == self.target or jug2 == self.target:
            print("reached")
            return True
        
        next_states = [
            (self.jug1, jug2), 
            (jug1, self.jug2), 
            (0, jug2),  
            (jug1, 0),
            (jug1 - min(jug1, self.jug2 - jug2), jug2 + min(jug1, self.jug2 - jug2)), 
            (jug1 + min(jug2, self.jug1 - jug1), jug2 - min(jug2, self.jug1 - jug1))  
        ]
        
        for next_state in next_states:
            if self.valid_state(next_state) and self.dfs(next_state, path):
                return True
        
        path.pop()
        return False
    
    def solve(self):
        print("Rules:")
        print("1. Fill Jug 1 completely.")
        print("2. Fill Jug 2 completely.")
        print("3. Empty Jug 1.")
        print("4. Empty Jug 2.")
        print("5. Pour water from Jug 1 to Jug 2 until Jug 2 is full or Jug 1 is empty.")
        print("6. Pour water from Jug 2 to Jug 1 until Jug 1 is full or Jug 2 is empty.")
        
        path = []
        if self.dfs((0, 0), path):
            print("Solution Path:")
            for step in path:
                print(step)
        else:
            print("No solution found.")

if __name__ == "__main__":
    jug1_capacity = 4
    jug2_capacity = 3
    target_amount = 2
    
    puzzle = waterjug(jug1_capacity, jug2_capacity, target_amount)
    puzzle.solve()
```
